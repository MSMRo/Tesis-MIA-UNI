## Análisis de variables categóricas

Analizar la distribución de las variables categóricas (si las hay) mediante conteos y visualizaciones (por ejemplo, gráficos de barras).


```python
import matplotlib.pyplot as plt
import seaborn as sns

class_counts = df['class'].value_counts()
display(class_counts)

plt.figure(figsize=(10, 6))
sns.barplot(x=class_counts.index, y=class_counts.values)
plt.title('Distribution of Class Categories')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Calidad: Verificación de tipos, rangos, duplicados y outliers
print("Tipos de datos:")
print(df.dtypes)

print("\nRangos de valores:")
for col in numeric_cols:
    print(f"{col}: Min={df[col].min()}, Max={df[col].max()}")

print("\nDuplicados:")
print(f"Número de filas duplicadas: {df.duplicated().sum()}")

print("\nOutliers:")
for col in numeric_cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f"{col}: {len(outliers)} outliers")

# Distribuciones y relaciones
print("\nDistribuciones y relaciones:")
for col in numeric_cols:
    plt.figure(figsize=(10, 5))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

print("\nBalance de la variable objetivo:")
sns.countplot(x='class', data=df)
plt.title('Balance de la variable objetivo')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()

print("\nCorrelaciones:")
correlation_matrix = df[numeric_cols].corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()

# Riesgos
print("\nRiesgos:")
print("Desbalance de clases:")
print(class_counts)

print("\nLeakage:")
print("No se puede determinar leakage sin más contexto sobre las variables.")

print("\nDrift:")
print("No se puede determinar drift sin datos temporales o de diferentes distribuciones.")

print("\nSesgos:")
print("No se puede determinar sesgos sin más contexto sobre los datos.")

# Conclusiones accionables
print("\nConclusiones accionables:")
print("1. Si se detectan outliers significativos, considerar su tratamiento (remoción o transformación).")
print("2. Si hay desbalance en la variable objetivo, considerar técnicas como sobremuestreo o submuestreo para balancear las clases.")
```
